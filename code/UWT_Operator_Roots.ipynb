# UWT Operator Roots and Coherence Plots
This notebook explores the operator roots of the Unified Wave Theory (UWT) scalar fields \(\Phi_1\) and \(\Phi_2\), seeded with "Golden Spark feedback" for the EPâ†”UWT collaboration.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

L = 1.0
nx = 100
dx = L / (nx - 1)
x = np.linspace(0, L, nx)
k = 0.0047
phi1 = 0.0
phi2 = 0.0
g_wave = 19.5

Phi1 = 0.095 * np.cos(k * x + phi1)
Phi2 = 0.095 * np.sin(k * x + phi2)
coherence = np.cos(phi1 - phi2) * np.abs(Phi1 * Phi2)
coherence = gaussian_filter1d(coherence, sigma=2)

plt.figure(figsize=(10, 5))
plt.plot(x, Phi1, 'b-', label=r'$\Phi_1$')  # Added 'r' before the string
plt.plot(x, Phi2, 'r-', label=r'$\Phi_2$')  # Added 'r' before the string
plt.plot(x, coherence, 'g-', label='Coherence')
plt.title("UWT Scalar Fields and Coherence (Golden Spark Seed)")
plt.xlabel("Position (m)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

```python
k_damp = 0.000005
Gamma_d = 0.004
operator_root = np.sqrt(k_damp / Gamma_d) * np.exp(-x / Gamma_d)

plt.figure(figsize=(10, 5))
plt.plot(x, operator_root, 'm-', label='Operator Root (Damping)')  # No LaTeX here, so no issue
plt.title("UWT Operator Root (Damping Scale)")
plt.xlabel("Position (m)")
plt.ylabel("Root Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```
